2. å°†å˜é‡ list å¡«å……åˆ°æ¯ä¸€åˆ—æ•°æ®ä¸­
3. éå†æ¯åˆ—ï¼Œå¦‚æœæ•°æ®åŒ…æ¶µäº†åˆ—åï¼Œè¿”å›1ï¼ŒæœªåŒ…å«ï¼Œè¿”å›0

```python
ctgs = [['purple', 'yellow',], ['purple', 'green'], ['yellow'], ['purple'], ['green'] ]
def dummy_var(lsts):
    var = set()
    for lst in lsts:
        for subctg in lst:
            var.add(subctg)
    sj = pd.DataFrame(np.random.randn(len(lsts), len(var)), columns = var)
    for col in sj.columns:
        sj[col] = lsts
        sj[col] = sj[col].map(lambda x:1 if col in x else 0)
    sj['variable'] = lsts 
    return sj
dummy_var(ctgs)
```

### 3. ä½¿ç”¨ sklearn åº“ä¸­çš„å…¶ä»–åˆ†ç±»æ–¹æ³•ï¼Œæ¥é¢„æµ‹ titanic çš„ç”Ÿå­˜æƒ…å†µ
ä¸Šå‘¨å·²ç”¨è¿‡ æ”¯æŒå‘é‡æœº SVM å’Œ éšæœºæ£®æ—ä¸¤ç§æ–¹æ³•ï¼Œä»£ç å¦‚ä¸‹

```python
# æ•°æ®å¯¼å…¥å’Œæ¸…ç†
train = pd.read_csv("train.csv") 
train = train.drop(['PassengerId','Name', 'Ticket', 'Cabin'], axis = 1)
train['Age'] = train['Age'].fillna(train['Age'].median())
train= train.dropna(how = 'any')

# è®¾ç«‹å…¬å¼ï¼Œå»ºç«‹æ•°æ®æ¡†
formula =  'Survived ~ C(Pclass) + C(Sex) + Age + SibSp' 
y, x = dmatrices(formula, data=train, return_type='matrix')

# å°†å…¬å¼æ•°æ®æ¡†ä¸­çš„æ•°æ®æ”¾å…¥ numpy array
# æŒ‘é€‰æ„Ÿå…´è¶£çš„å˜é‡æ¥åˆ†æ:é€šè¿‡åˆšæ‰çš„å›å½’åˆ†æï¼Œå·²çŸ¥æ€§åˆ«å’Œèˆ¹èˆ±å¯¹æ˜¯å¦ç”Ÿè¿˜çš„å½±å“è¾ƒå¤§ï¼Œè¿™é‡Œé€‰æ‹©ä»£è¡¨æ€§åˆ« 3 å’Œèˆ¹èˆ± 2 çš„ä¸¤åˆ—æ¥åˆ†æ
X = np.asarray(x)
# y å€¼æ˜¯è¿™æ ·çš„æ•°æ®ï¼Œarray([[ 0.], [ 1.], [ 1.], [ 1.] ...]), ç”¨ np.ravel() è½¬æ¢ä¸º array([ 0.,  1.,  1.,  1.,  0.,  0., ...])
y = np.asarray(y).ravel()
X = X[:,[2, 3]]  

# éšæœºæ’åˆ— x, yå€¼ 
np.random.seed(123)
order = np.random.permutation(len(x))
X = X[order]
y = y[order].astype(np.float)

# åšä¸€ä¸ªäº¤å‰éªŒè¯ï¼Œå‰80%çš„æ•°æ®ç”¨æ¥è®­ç»ƒï¼Œå20%çš„æ•°æ®ç”¨æ¥æ£€éªŒ
X_train = X[:int(.8 * len(x))]
y_train = y[:int(.8 * len(x))]
X_test = X[int(.8 * len(x)):]
y_test = y[int(.8 * len(x)):]

# 1. æ”¯æŒå‘é‡æœº
# gamma å‚æ•°: å‚æ•° gamma è¡¨ç¤ºè¶…å¹³é¢çš„çº¿æ€§å¹³æ»‘åº¦ï¼Œå½“ä½¿ç”¨çº¿æ€§æ ¸å‡½æ•°æ„å»º SVM æ—¶ï¼Œæ¨¡å‹ä¸­ä¸å­˜åœ¨ gamma å‚æ•°ã€‚
# æˆ‘ä»¬çŸ¥é“: gamma è¶Šå°ï¼Œè¶…å¹³é¢è¶Šæ¥è¿‘äºç›´çº¿ï¼Œä½†æ˜¯å¦‚æœ gamma é€‰å–è¿‡å¤§ï¼Œè¶…å¹³é¢å°†ä¼šå˜å¾—éå¸¸å¼¯æ›²ï¼Œè¿™ä¼šå¯¼è‡´è¿‡åº¦æ‹Ÿåˆé—®é¢˜ã€‚
poly = SVC(kernel='poly', gamma=2.5)
poly.fit(X_train, y_train)
print u'1. æ”¯æŒå‘é‡æœºç®—æ³•é¢„æµ‹è¯„åˆ† :',poly.score(X_test, y_test) 
```

### 5. ç”¨ numpy åº“å®ç° K-nearest neighbors åˆ†ç±»
#### (1) ä»€ä¹ˆæ˜¯ KNN ç®—æ³•ï¼Ÿ
- K Nearest Neighborç®—æ³•åˆå«KNNç®—æ³•ï¼Œæ˜¯æœºå™¨å­¦ä¹ é‡Œé¢ä¸€ä¸ªæ¯”è¾ƒç»å…¸ä¸”å®¹æ˜“ç†è§£çš„ç®—æ³•ã€‚å…¶ä¸­çš„ K è¡¨ç¤ºæœ€æ¥è¿‘è‡ªå·±çš„ K ä¸ªæ•°æ®æ ·æœ¬ã€‚
- KNN ç®—æ³•å’Œ K-Means ç®—æ³•ä¸åŒçš„æ˜¯ï¼ŒK-Means ç®—æ³•ç”¨æ¥èšç±»ï¼Œç”¨æ¥åˆ¤æ–­å“ªäº›ä¸œè¥¿æ˜¯ä¸€ä¸ªæ¯”è¾ƒç›¸è¿‘çš„ç±»å‹ï¼Œè€Œ KNN ç®—æ³•æ˜¯ç”¨æ¥åšå½’ç±»çš„ï¼Œä¹Ÿå°±æ˜¯è¯´ï¼Œæœ‰ä¸€ä¸ªæ ·æœ¬ç©ºé—´é‡Œçš„æ ·æœ¬åˆ†æˆå¾ˆå‡ ä¸ªç±»å‹ï¼Œç„¶åï¼Œç»™å®šä¸€ä¸ªå¾…åˆ†ç±»çš„æ•°æ®ï¼Œé€šè¿‡è®¡ç®—æ¥è¿‘è‡ªå·±æœ€è¿‘çš„ K ä¸ªæ ·æœ¬æ¥åˆ¤æ–­è¿™ä¸ªå¾…åˆ†ç±»æ•°æ®å±äºå“ªä¸ªåˆ†ç±»ã€‚
- å¯ä»¥ç®€å•å°† KNN ç®—æ³•ç†è§£ä¸ºï¼Œæ•°æ®ç‚¹ A å±äºå“ªä¸€ç±»ï¼Œç”±ç¦»å®ƒæœ€è¿‘çš„Kä¸ªç‚¹æ¥æŠ•ç¥¨ï¼ŒæŠ•ç¥¨ç»“æœå†³å®š A è¯¥è¢«å½’ä¸ºå“ªä¸€ç±»

#### ä¸€ä¸ªä¾‹å­
<img src="http://coolshell.cn//wp-content/uploads/2012/08/220px-KnnClassification.svg_.png" width="27%" height="100%" style="float:left"/>

#### å›¾ä¸­çš„æœ‰ä¸¤ç±»æ•°æ®ï¼Œæ­£æ–¹å½¢å’Œä¸‰è§’å½¢ï¼›åœ†å½¢ä¸ºå¾…åˆ†ç±»çš„æ•°æ®ã€‚
- K=3ï¼Œç¦»åœ†å½¢ç‚¹æœ€è¿‘çš„æœ‰2ä¸ªä¸‰è§’å½¢å’Œ1ä¸ªæ­£æ–¹å½¢ï¼Œåˆ™ç”±è¿™3ä¸ªç‚¹æŠ•ç¥¨ï¼Œäºæ˜¯åœ†å½¢çš„å¾…åˆ†ç±»ç‚¹å±äºä¸‰è§’å½¢
- K=5ï¼Œé‚£ä¹ˆç¦»åœ†å½¢ç‚¹æœ€è¿‘çš„æœ‰2ä¸ªä¸‰è§’å½¢å’Œ3ä¸ªæ­£æ–¹å½¢ï¼Œç”±5ä¸ªç‚¹æŠ•ç¥¨ï¼Œäºæ˜¯åœ†å½¢çš„å¾…åˆ†ç±»ç‚¹å±äºæ­£æ–¹å½¢ã€‚

#### (2) é€‰å–äº† iris æ•°æ®æ¥è¿›è¡Œ KNN åˆ†ç±»

```python
# å¯¼å…¥æ•°æ®ï¼Œç»å…¸èŠ±åˆ†ç±»æ•°æ®é›†
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.6, random_state=1)
 
# äº¤å‰éªŒè¯
train = np.array(zip(X_train,y_train))
test = np.array(zip(X_test, y_test))
 
# 1) å·²çŸ¥ä¸¤ä¸ªæ•°æ®ç‚¹ï¼Œè®¡ç®—æ¬§æ°è·ç¦»
def get_distance(point1, point2):
    points = zip(point1, point2)
    diffs_squared_distance = [pow(a - b, 2) for (a, b) in points]
    return math.sqrt(sum(diffs_squared_distance))

def get_neighbours(training_set, test_instance, k):
    distances = [_get_tuple_distance(training_instance, test_instance) for training_instance in training_set]
 
    # 1 ä¸ºè®­ç»ƒå€¼ä¸æ£€éªŒå€¼ä¹‹é—´çš„è·ç¦» 
    sorted_distances = sorted(distances, key=itemgetter(1))
 
    # åªæŠ½å–è®­ç»ƒå€¼
    sorted_training_instances = [tuple[0] for tuple in sorted_distances]
 
    # é€‰æ‹© k ä¸ªå…ƒç´ 
    return sorted_training_instances[:k]
 
def _get_tuple_distance(training_instance, test_instance):
    return (training_instance, get_distance(test_instance, training_instance[0]))
 
def get_majority_vote(neighbours):
    # index 1 is the class
    classes = [neighbour[1] for neighbour in neighbours]
    count = Counter(classes)
    return count.most_common()[0][0]

def main():
 
    predictions = []
 
    # å…ˆéšæ„å°†kè®¾ç½®ä¸º5ï¼Œå³åˆå§‹å°è¯•çš„åˆ†ç±»å€¼ä¸º5
    k = 5
 
    for x in range(len(X_test)):
 
            print 'Classifying test instance number ' + str(x) + ":",
            neighbours = get_neighbours(training_set=train, test_instance=test[x][0], k=5)
            majority_vote = get_majority_vote(neighbours)
            predictions.append(majority_vote)
            print 'Predicted label=' + str(majority_vote) + ', Actual label=' + str(test[x][1])
 
    # æ€»ç»“
    print '\nThe overall accuracy of the model is: ' + str(accuracy_score(y_test, predictions)) + "\n"
    report = classification_report(y_test, predictions, target_names = iris.target_names)
    print 'A detailed classification report: \n\n' + report
 
if __name__ == "__main__":
    main()
```
