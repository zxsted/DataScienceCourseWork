## 1. Áà¨Âèñ‰∏äËØÅÁªºÊåáÂéÜÂè≤ËÇ°‰ª∑
- Áà¨Âèñ SSE Êï∞ÊçÆ (000001.SS) : pandas_datareader.data  

#### ÂèØËßÜÂåñÁªìÊûú
‰∏äËØÅÁªºÊåáÂàÜÂà´Âú® 2007 Âπ¥Âíå 2015 Âπ¥ÁªèÂéÜ‰∫Ü‰∏§Ê¨°üêÇÂ∏Ç

```python
%matplotlib inline 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
import pandas_datareader.data as web
from datetime import *
import numpy as np
start = datetime(2006, 3, 18)
end = datetime(2016, 5, 26)
f = web.DataReader('000001.SS', 'yahoo', start, end)
f.head()
```

```python
f['Close'].plot(color='lightblue', legend=None, figsize=(9,6))
plt.ylabel('Close')
plt.legend()
plt.show()
```

## 2. ËÆ°ÁÆóÊ∂®Ë∑åÂπÖÂπ∂ÂàÜÊûê

### ËÆ°ÁÆóÊúàÊ∂®Ë∑å
Â∞ÜÊï∞ÊçÆÊåâÊúà‰ªΩÊéíÂàóÔºåÂπ∂ËÆ°ÁÆóÊúàÊúÄÂ§ß‰∏éÊúÄÂ∞èÂÄºÁöÑÂ∑ÆÂÄº

```python
f.index
mon_mean = f.groupby(pd.TimeGrouper('M')). \
agg({'Close':'mean'})
mon_change = mon_mean[1:]-mon_mean[:-1].values
mon_change[0:3]
```

#### 06Âà∞16ÂçÅÂπ¥Èó¥ÊúàÂπ≥ÂùáÂèòÂåñÂõæ
07Âπ¥ÁâõÂ∏ÇÊúàÂ¢ûÈïøÈÄüÂ∫¶ÊØî15Âπ¥ÁâõÂ∏ÇÊõ¥Âø´ÔºåÁÑ∂ËÄåÈöèÂêé......

```python
mon_change['Close'].plot.area(stacked=False, figsize=(9,6))
plt.legend()
```

### ËÆ°ÁÆóÊó•Ê∂®Ë∑å
#### ËÆ°ÁÆóÊó•Ê∂®Ë∑åÂπÖÁöÑ‰∏§‰∏™ÊñπÊ≥ï
- pandasËá™Â∏¶Ê∂®Ë∑åÂπÖËÆ°ÁÆóÂáΩÊï∞ pct_change()
- ‰ª•ÂèäÂü∫‰∫éDataFrameÁâπË¥®ÁöÑ‰∏Ä‰∏™Á•ûÂ•áÊñπÊ≥ï

```python
f.pct_change()[0:3]
```

```python
(f[1:] - f[:-1].values)[0:2]
```

#### Â¢ûÂä†pct_ChangeÂíåChangeÂàóÔºåÂêåÊó∂ÂéªÈô§NA

```python
f['pct_Change'] = f.pct_change()['Close']
f['Change'] = (f[1:] - f[:-1].values)['Open']
f.fillna(0, inplace=True)
```

```python
f[0:2]
```

### ÊäΩÂèñ‰∏ÄÊÆµÊó∂Èó¥Êù•ÁúãÊØèÊó•Ê∂®Ë∑åÂπÖ

#### ÊäΩÂèñ2015Âπ¥1ÊúàÂà∞9ÊúàËøôÊÆµÊó∂Èó¥ÁöÑÊï∞ÊçÆÔºåËé∑ÂèñÊ∂®Ë∑å‰∏§‰∏™Â≠êÈõÜ

```python
sub1year = f.loc['2015-1':'2015-9'].reset_index()
sub1yearup = sub1year[sub1year['Change'].apply(lambda x: x>0)]
sub1yeardn = sub1year[sub1year['Change'].apply(lambda x: x<=0)]
```

#### ÁîªstemÂõæÔºåÊåâÁÖßÂõΩÂÜÖ‰π†ÊÉØÔºåÁ∫¢Ëâ≤‰∏∫Ê∂®ÔºåÁªøËâ≤‰∏∫Ë∑å
#### ÂèØËßÜÂåñÁªìÊûú
2015Âπ¥2ÊúàÂà∞5ÊúàÂ∫ïËøôÊÆµÊó∂Èó¥ÂÜÖÔºå‰∏äËØÅÁªºÊåáÁªèÂéÜ‰∫ÜÊØîËæÉÂØÜÈõÜÁöÑÊ∂®ÂπÖ„ÄÇ   
2015Âπ¥5ÊúàÂ∫ïÂêéÂà∞8Êúà‰πãÈó¥ÔºåËÇ°‰ª∑ÁªèÂéÜ‰∫ÜÂ§ßÂπÖÂ∫¶ÁöÑË∑≥Ê∞¥......

```python
plt.figure(figsize=(10,7))
up = plt.stem(sub1yearup['Date'],sub1yearup['Change'].values,'-.')
dn = plt.stem(sub1yeardn['Date'],sub1yeardn['Change'].values,'-.')
plt.setp(up, 'markerfacecolor', 'crimson')
plt.setp(dn, 'markerfacecolor', 'c')
plt.show()
```

### 3. ‰∏äËØÅÁªºÊåá‰∏éÈÅìÁêºÊñØÂ∑•‰∏öÊåáÊï∞
ÊäìÂèñÈÅìÁêºÊñØÂ∑•‰∏öÊåáÊï∞Êï∞ÊçÆ

```python
start = datetime(2006, 3, 18)
end = datetime(2016, 5, 26)
d = web.DataReader('^DJI', 'yahoo', start, end)
d.head()
```

#### ÂêàÂπ∂DJI‰∏éSSEÊï∞ÊçÆ

```python
compare = pd.DataFrame({'SSE':f['Close'], 'DJI':d['Close']})
compare.fillna(0, inplace=True)
compare[0:5]
```

#### ÁîªÂõæÊØîËæÉDJI‰∏éSSE
ÈÉΩÂú®07Âπ¥Êú´Âíå15Âπ¥ÂàùÁªèÂéÜ‰∫ÜÂ≥∞ÂÄºÔºåÂêåÊ†∑Âú®09Âπ¥Â§ßÂπÖ‰∏ãË∑å

```python
compare.plot(figsize=(9,6))
```

#### ‰∏çËøáÔºå06Âà∞16Âπ¥Èó¥ÔºåSSE‰∏éDJIÁöÑÊ∂®Ë∑åÊòØÂê¶Áõ∏ÂÖ≥Âë¢ÔºüÂÜçÁîª‰∏™ÂõæÁúãÁúã
Âú®Êï∞ÊçÆË°®‰∏≠Âπ∂ÂÖ•Ê∂®Ë∑åÂÄº

```python
compare['dpct_Change'] = d.pct_change()['Close']
compare['fpct_Change'] = f['pct_Change']
compare.fillna(0, inplace=True)
compare[0:5]
```

#### Ëá≥Â∞ë‰ªéÁõ∏ÂÖ≥Á≥ªÊï∞Êù•ÁúãÂë¢Ôºå10Âπ¥Êù•‰ªñ‰ø©ÁöÑÊ∂®Ë∑åÂπ∂Ê≤°ÊúâÊ£Æ‰πàÂÖ≥Á≥ªÔºåÁõ∏ÂÖ≥Á≥ªÊï∞ÈùûÂ∏∏‰Ωé

```python
import seaborn as sns
sns.jointplot("dpct_Change","fpct_Change",compare, \
              kind="scatter", color='forestgreen', alpha = 0.5)
```
